```python
def create_db(test_db, test_file, table_name):
    con = sql3.connect(test_db)
    cursor=con.cursor()

    df = pd.read_csv(test_file, encoding='unicode_escape')
    df.to_sql(table_name , con , if_exists='replace' , index=False)

    results = cursor.execute(f"SELECT * FROM {table_name}").fetchall()

    con.commit()
    con.close()

    return results
```

```python
create_table = """CREATE TABLE Physics_Laureates(
                        Id INTEGER PRIMARY KEY,
                        Firstname TEXT,
                        Surname TEXT,
                        Born DATE,	
                        Birth_Location VARCHAR(50),
                        Noble_recieved DATE,
                        Motivation TEXT,
                        Organization_name TEXT,
                        Organisation_Location TEXT
);
                """

cursor.execute(create_table)
```

```python
laureates = create_db("Noble Prize - Laureates.db", "noble_laureates_final.csv", "Physics_Laureates")
```
